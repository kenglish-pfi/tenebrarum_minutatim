# Count Phoneme Transitions

Reads in the CMU pronunciation dictionary

http://svn.code.sf.net/p/cmusphinx/code/trunk/cmudict/cmudict-0.7b

Simply counts transitions pair-wise and sorts

```python
import codecs
import io
import pandas as pd
import sys
```

```python
pd.set_option('display.max_rows', None)
```

```python
def loadCmuDict(file):
    D = {}
    with codecs.open(file, encoding="iso-8859-1") as f:
        for line in f:
            line = line.rstrip()
            if line[0:3] == ';;;':
                continue
            A = line.split(' ')
            # First token is term, rest of tokens are pronounciation
            D[A[0]] = [t for t in A[1:] if t != ""]
    return D
```

```python
def computePhonemePairs(cmu_dict):
    D = {}
    for term in cmu_dict:
        n = len(cmu_dict[term])
        if n < 2:
            continue
        for i in range(n - 1):
            pair = " ".join(cmu_dict[term][i:i+2])
            if pair not in D:
                D[pair] = 0
            D[pair] += 1
    df = pd.DataFrame(list(D.items()), columns=["pair", "count"]).sort_values("count", ascending=False)
    return df
#   
```

```python
def main():
    return computePhonemePairs(loadCmuDict("/data/repo/svn.code.sf.net/p/cmusphinx/cmudict/cmudict-0.7b"))
```

```python
df_phoneme_pairs = main()
df_phoneme_pairs
```
