```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sin4(X):
    list0 = []
    list1 = []
    list2 = []
    list3 = []
    for x in X:
        list0.append( math.sin(x) )
        list1.append( math.sin(math.pi/2.0 - x) )      # note: negative, sin(-x) = -sin(x)
        list2.append( math.sin(math.pi + x) )
        list3.append( math.sin(3.0*math.pi/2.0 - x) )
    return list0, list1, list2, list3
```

```python
def pltSin4():
    # generate some X values from 0 .. 2Ï€
    X = np.array(range(100)) * 2 * math.pi / 100.0
    Y0, Y1, Y2, Y3 = sin4(X)
    plt.plot(X, Y0, c='tab:blue')
    plt.plot(X, Y1, c='tab:orange')
    plt.plot(X, Y2, c='tab:green')
    plt.plot(X, Y3, c='tab:red')
    plt.show()
```

```python
pltSin4()
```

```python
def cos4(X):
    list0 = []
    list1 = []
    list2 = []
    list3 = []
    for x in X:
        list0.append( math.cos(x) )
        list1.append( math.cos(math.pi/2.0 + x) )     # note: all positive ... cos(-x) = cos(x)
        list2.append( math.cos(math.pi + x) )
        list3.append( math.cos(3.0*math.pi/2.0 + x) )
    return list0, list1, list2, list3
```

```python
def pltCos4():
    X = np.array(range(100)) * 2 * math.pi / 100.0
    Y0, Y1, Y2, Y3 = cos4(X)
    # rotated colors so sin and cos plots look the same
    plt.plot(X, Y0, c='tab:orange')  # was second color  âˆ´  d/dx[ sin(x) ] = cos(x) = sin(math.pi/2.0 - x)
    plt.plot(X, Y1, c='tab:green')   # â†‘
    plt.plot(X, Y2, c='tab:red')     # was last color
    plt.plot(X, Y3, c='tab:blue')    # was first color
    plt.show()
```

```python
pltCos4()
```
