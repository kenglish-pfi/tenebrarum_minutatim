```python
def sct_to_fraction(sc_tup):
    '''
    Convert a Shortcut Collatz Tuple to a reduced fraction
    '''
    p2, p3, c = sc_tup
    fract = Fraction(2**p2 - c, 3**p3)
    return (fract.numerator, fract.denominator)
#

def filter_int(sc_fraction_list):
    for fract in sc_fraction_list:
        if fract[1] == 1:
            yield fract[0]
#

# tup expresses a rational number of the form:
#Â  (2**p2 - c) / (3**p3)


def save_top_generations(max_generation):
    with open("generation_%d"%(max_generation-1), "w") as f0:
        with open("generation_%d"%(max_generation), "w") as f1:
        
            def seq_next(generation, sc_tup):
                '''
                Generate 2 tuples for the next generation from a single tuple from the current generation
                '''
                p2, p3, c = sc_tup

                t1 = (p2 + 1, p3, c)
                t = t1
                if generation == max_generation:
                    (numerator, denominator) = sct_to_fraction(t)
                    if denominator == 1:
                        f1.write("%d\t0\t%d\t%d\t%d\n"%(numerator, *t))
                elif generation == max_generation-1:
                    (numerator, denominator) = sct_to_fraction(t)
                    if denominator == 1:
                        f0.write("%d\t0\t%d\t%d\t%d\n"%(numerator, *t))
                    seq_next(generation+1, t)
                else:
                    seq_next(generation+1, t)

                t2 = (p2 + 1, p3 + 1, c*3 + 2**p2)
                t = t2
                if generation == max_generation:
                    (numerator, denominator) = sct_to_fraction(t)
                    if denominator == 1:
                        f1.write("%d\t1\t%d\t%d\t%d\n"%(numerator, *t))
                elif generation == max_generation-1:
                    (numerator, denominator) = sct_to_fraction(t)
                    if denominator == 1:
                        f0.write("%d\t1\t%d\t%d\t%d\n"%(numerator, *t))
                    seq_next(generation+1, t)
                else:
                    seq_next(generation+1, t)

            #
            seq_next(1, (0,0,0))
#
```
